Kita dapat melakukan konversi tipe data bawaan dengan menggunakan fungsi konversi tipe bawaan (standard type) misalnya: int(), float(), str(), dll.

Misalnya tipe int() diubah menjadi float()

```python
3
```

```python
type(3)
```

```python
float(3)
```

```python
print(type(float(3)))
```

Konversi float ke int akan bersifat flooring/truncanting atau menghilangkan nilai di belakang koma.

```python
5.2
```

```python
type(5.2)
```

```python
int(5.2)
```
