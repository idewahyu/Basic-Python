#Variabel
- Variables are containers for storing data values.
- Python does not have a command to declare variables.
- A variable is created when you first assign a value to it.
- Variables don't need to be declared of a specific type, and can even be retyped once defined.
- To assign a value to a variable, use the "=" operator, between the variable and the value we want to store.

```python
x = 3
```

```python
y = 5.0
```

```python
z = "final letter of alphabet"
```

#How to insert the value of variable in string

There are several ways to enter a variable value in a string:
1. Directly concatenate variables in the print() statement.

```python
X = 3
type(X)
```

```python
print("The value of X is", X)
```

2. To display text (string), you can use a format string.

```python
print("My ID card number is {}".format("112233"))
```

3. Using operator "%" added with "argument specifiers".

  Some examples of commonly used argument specifiers:

  - %s - String
  - %d - Integers
  - %f - Decimal Number

```python
Name = "Zachary Eisen"
```

```python
Age = 25
```

```python
print("The age of %s is %d years old" % (Name, Age))
```
