#Dynamic Typing
Python is known as one of the languages that implement dynamic typing, which means the Python interpreter does type checking only as code runs, and the type of a variable is allowed to change over its lifetime. Here a couple of dynamic typing examples:

- type()

  Type() is one of the Python's built-in functions that helps you find the class type of the variable given as input.

```python
a = 100
type(a)
```
