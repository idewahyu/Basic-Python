#IF
- In Python, any boolean expression evaluating to True or False appears after the if keyword.
- If the expression evaluates to True, the statements inside the if statement will be executed.
- If the expression evaluates to False, the next block (after the IF statement) will be executed.
- IF statement has an indentation (indentation) after the colon (:).

```python
keyword = "clavis"

if keyword:
    print("The correct keyword is {}".format(keyword))
```

#Else
- The else statement can be combined with the if statement. 
- It contains a block of code that is executed if the conditional expression in the if statement becomes 0 or the value is FALSE.
- The else statement is an optional statement and there could be at most only one else statement following if.


```python
age = int(input("enter your age: "))

if age >= 17:
    print("Already have an ID Card")
else:
    print("Don't have an ID Card")
```
