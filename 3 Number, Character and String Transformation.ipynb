#Uppercase and Lowercase

- upper()

  upper() is a built-in method used for string handling. The upper() methods returns the uppercased string from the given string. It converts all lowercase characters to uppercase. If no lowercase characters exist, it returns the original string.

```python
your_name = "claudia hartmann"
your_name.upper()
```

- lower()

  lower() is a built-in method used for string handling. The lower() methods returns the lowercased string from the given string. It converts all uppercase characters to lowercase. If no uppercase characters exist, it returns the original string.

```python
your_name = "CLAUDIA HARTMANN"
your_name.lower()
```

#Prefix and Suffix
1. rstrip()
  
  rstrip() can remove any white spaces at the end of the string.

```python
a = "Data    "
a
```

```python
a.rstrip()
```

2. lstrip()
  
  lstrip() can remove spaces to the left of the string.

```python
a = "   Data"
a
```

```python
a.lstrip()
```

3. strip()

  string() can remove spaces at the beginning and at the end of the string

```python
a = "   Data   "
a
```

```python
a.strip()
```

#startswith() & endswith()
- startswith()

  The startwith() method will return True if the string starts with the word we want, otherwise it will return False.

```python
a = "Data Science"
a.startswith("Science")
```

```python
b = "Data Science Track"
b.startswith("Data")
```

```python
c = "Data Science Track"
c.startswith("Data Science")
```

```python
d = "Data Science Track"
d.startswith("Track")
```

- endswith()

  Endswith() method is the opposite of the startswith() method, this method will return True if the string ends with the specified suffix, otherwise it will return False.

```python
a = "Data Science"
a.endswith("Science")
```

```python
e = ("Data Scientist Data Engineer")
e.startswith("Data Engineer")
```
